## 💳 Financial Features

```python
df["monthly_income_avg"] = np.random.randint(8000, 60000, len(df))
df["annual_income"] = df["monthly_income_avg"] * 12

df["income_variance"] = np.random.uniform(0.1, 0.7, len(df))
```

```python
df.head(10)
```

## 📊 Seasonality & Event Features

```python
# simulate seasonal workers
seasonal_flag = np.random.choice([0,1], size=len(df), p=[0.6,0.4])

df["seasonality_score"] = np.where(seasonal_flag==1,
                                  np.random.uniform(0.6,1.0,len(df)),
                                  np.random.uniform(0.0,0.4,len(df)))

df["income_concentration"] = np.where(seasonal_flag==1,
                                      np.random.uniform(0.5,0.9,len(df)),
                                      np.random.uniform(0.1,0.4,len(df)))

df["event_dependence_score"] = df["seasonality_score"]
```

```python
seasonal_features = [
    'seasonality_score',
    'income_concentration',
    'event_dependence_score'
]
df[seasonal_features].head(10)
```

## 🧾 Behavioral Features

```python
df["bill_payment_score"] = np.random.uniform(0,1,len(df))
df["upi_txn_count"] = np.random.randint(10,300,len(df))

df["ecommerce_spend"] = np.random.randint(500,20000,len(df))
df["ecommerce_spend_ratio"] = df["ecommerce_spend"] / df["monthly_income_avg"]
```

```python
behavioral_features = [
    'bill_payment_score',
    'upi_txn_count',
    'ecommerce_spend',
    'ecommerce_spend_ratio'
]
df[behavioral_features].head(10)
```

# 🎯 Target

```python
risk = (
    0.20 * df["income_variance"] * (1 - df["seasonality_score"]) +
    0.20 * (1 - df["bill_payment_score"]) +
    0.15 * (1 - df["location_stability"]) +
    # 0.15 * (1 - df["questionnaire_score"]) +
    0.10 * df["ecommerce_spend_ratio"] +
    0.10 * (1 - df["job_stability_score"]) +
    0.10 * (df["income_concentration"] * (1 - df["event_dependence_score"]))
)

df["default"] = (risk > 0.5).astype(int)
```
